```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
```

```python
path = "../../data/telecomunicatii.csv"

df= pd.read_csv(path)
df.head()
```

```python
print(df['area_code'].unique())
```

```python
print(df.isnull().sum())
```

```python
X = df.drop(columns=['payment_delay']) 
y = df['payment_delay'] 
```

```python
le = LabelEncoder()
y =le.fit_transform(y)

categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
print(categorical_cols)
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(numeric_cols)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough' # keep numerical col 
)

X_processed = preprocessor.fit_transform(X)

encoded_cols = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)
all_features = list(encoded_cols) + numeric_cols

X_df = pd.DataFrame(X_processed, columns=all_features)
```

```python
X_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.2, random_state=42, stratify=y)

print(X_train.shape)
print(X_test.shape)
```

```python
scaler = StandardScaler()

X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
X_train
```

```python
y_train
```

```python
X_train_df = X_train.copy()
X_test_df = X_test.copy()

y_train_df = pd.DataFrame(y_train, columns=['Risk'])
y_test_df = pd.DataFrame(y_test, columns=['Risk'])

path = "/home/dana/partera_roz/data/preprocessed_data"
X_train_df.to_csv('/home/dana/partera_roz/data/preprocessed_data/X_train_df.csv', index=False)
X_test_df.to_csv('/home/dana/partera_roz/data/preprocessed_data/X_test_df.csv', index=False)

y_train_df.to_csv('/home/dana/partera_roz/data/preprocessed_data/y_train_df.csv', index=False)
y_test_df.to_csv('/home/dana/partera_roz/data/preprocessed_data/y_test_df.csv', index=False)
```
